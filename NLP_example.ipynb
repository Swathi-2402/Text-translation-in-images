```python
from googletrans import Translator

def translate_sanskrit(text, target_language='en'):
    translator = Translator()
    translation = translator.translate(text, dest=target_language)
    return translation.text

sanskrit_text = "नमस्ते, कृपया ध्यान दें कि यह केवल एक उदाहरण है।"
translated_text = translate_sanskrit(sanskrit_text)

print(f"Sanskrit Text: {sanskrit_text}")
print(f"Translated Text: {translated_text}")
```
